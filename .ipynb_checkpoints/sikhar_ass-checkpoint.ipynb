# f)	Test Null No Significant difference in Average Inflation..consumer.prices..annual and GDP.growth..annual. interpret based on p-value

```python
from scipy.stats import ttest_ind
```

```python
data.head(2)
```

```python
inflation=data.loc[data['Series Name']=='Inflation, consumer prices (annual %)'].select_dtypes(include=['number'])
inflation
```

```python
melted_df = inflation.melt(var_name='Year', value_name='Value')
melted_df['Year'] = melted_df['Year'].str.extract('(\d+)', expand=False).astype(int)
melted_df
```

```python
gdp_growth =data.loc[data['Series Name']=='GDP per capita growth (annual %)'].select_dtypes(include=['number'])
gdp_growth
melted_df_2 = gdp_growth.melt(var_name='Year', value_name='Value')
melted_df_2['Year'] = melted_df_2['Year'].str.extract('(\d+)', expand=False).astype(int)
melted_df_2
```

```python


# Perform t-test
t_statistic, p_value = ttest_ind(melted_df, melted_df_2)

print(p_value[1])
# Interpret the result
if p_value[1] > 0.05:
    print("There is no significant difference in the average inflation and GDP growth (p > 0.05)")
else:
    print("There is a significant difference in the average inflation and GDP growth (p <= 0.05)")
```

# g)	Test Null No Significant difference in Avearge Population.growth..annual, Inflation, consumer prices (annual %) and GNI.growth..annual... . interpret based on p-value

```python
p_values = {}
variables = data.columns[4:]
for i in range(len(variables)):
    for j in range(i+1, len(variables)):
        var1 = variables[i]
        var2 = variables[j]
        _, p_value = ttest_ind(data[var1], data[var2])
        p_values[(var1, var2)] = p_value

# Interpret p-values
for (var1, var2), p_value in p_values.items():
    if p_value > 0.05:  # If p-value > 0.05, fail to reject the null hypothesis
        print(f"No significant difference between {var1} and {var2}: p-value = {p_value}")
    else:
        print(f"Significant difference between {var1} and {var2}: p-value = {p_value}")
```

# h)	Univariate Time Series Forecasting, Stationarity Test, Decomposition, ARIMA Modelling  and Interpretation for variables
1)	GDP growth (annual %)
2)	Inflation, consumer prices (annual %)


```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import pandas as pd

# Define a function for the stationarity test
def stationarity_test(data):
    result = adfuller(data)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f' {key}: {value}')

# Assuming gdp_growth is your DataFrame column containing GDP growth data
# Convert gdp_growth to a Series if it's a DataFrame
gdp_growth_series = gdp_growth.squeeze()

# Perform stationarity test
stationarity_test(gdp_growth_series)

# Decomposition
decomposition = seasonal_decompose(gdp_growth_series, model='additive', period=1)
```
